`````python
# building execution time of every functions
def function_exec_time(funct):
  def wrapper(*args, **kwargs):
    start = time.time()
    result = funct(*args, **kwargs)
    end = time.time()
    execution_time = end - start
    print(f"execution time of this function is : {(end - start):.4f} s")
    return result
  return wrapper
`````

`````python
# fonction de localisation
@function_exec_time
def location(
             x: str,
             key
             ):

  """
  location_type can be lon or lat
  if you want longitude, function input is lon
  if you want latitude, function input is lat
  Adding the key is strongly couraged
  """
  # intanciation de la classe Nominatim
  geolocat = Nominatim(user_agent="solarpanels_app")

  key = key

  if key == "lon":
    return geolocat.geocode(x).longitude

  elif key == "lat":
    return geolocat.geocode(x).latitude

  else:
    print("enter a right key, 'either' lon or 'lat', not another key")
`````

#### **Traitement avec la fonction pandas_udf.**

`````python
from pyspark.sql.functions import pandas_udf
from pyspark.sql.types import StringType, DoubleType


# fonction de localisation
@function_exec_time
def location(
             ville: str,
             coord_type
             ):

  """
  location_type can be lon or lat
  if you want longitude, function input is lon in lieu of coor_type
  if you want latitude, function input is lat in lieu of coor_type
  Adding the key is strongly couraged
  """
  # intanciation de la classe Nominatim
  geolocat = Nominatim(user_agent="solarpanels_app")

  key = coord_type

  if key == "lon":
    return geolocat.geocode(ville).longitude

  elif key == "lat":
    return geolocat.geocode(ville).latitude

  else:
    print("enter a right coord_type, 'either' lon or 'lat', not another key")


# Création de la pandas_udf pour obtenir la longitude
# renvoyant des décimaux

# villes is columns ville of the dataframe

@function_exec_time
@pandas_udf(DoubleType())
def get_location_lon(villes):
    return villes.apply(lambda x: location(x, "lon"))


# Création de la pandas_udf pour obtenir la latitude
# Renvoyant des décimaux
@function_exec_time
@pandas_udf(DoubleType())
def get_location_lat(villes):
    return villes.apply(lambda x: location(x, "lat"))
`````

`````python
# constitution des colonnes du dataframe finale
col_name = list(city_table_response1.keys())
col_name.insert(0, "ville")
col_name.insert(0, "parameters")
print(col_name)
`````

`````python
# rows by a point(here a city) defined with latitude and longitude
# and any parameters it has
@function_exec_time
def POWER_Monthly_Annual_city_data(
    start: int,
    end: int,
    city: str,
    latitude: float,
    longitude: float,
    parameters: str,
    columns: list
                              ):

  """
  give the parameters and parameters value of a city each month a year
  from start year to end year
  """

  sub_data_collect = pd.DataFrame(columns=columns)


  for i in parameters:
    params = {
            "start": start,
            "end": end,
            'latitude': latitude,
            "longitude": longitude,
            "community": "re", # re as Renewable Energy,
                               # sb as Sustainable Buildings
                               # ag as AGroclimatology
            "parameters": i,
            "format": "json",
            "user": None,
            "header": "False",
            "site-elevation": None,
            "wind-elevation": None,
            "wind-surface": None}


    city_table_response = requests.get(POWER_Monthly_Annual_API, params=params)
    city_table_response = city_table_response.json()
    city_table_response = city_table_response.get('properties')['parameter'] \
                                                                [i]

    new_rows = pd.Series(city_table_response.values())
    new_data = insert_city_parameter(new_rows, city, i)
    new_data.columns = columns
    sub_data_collect = pd.concat([sub_data_collect, new_data], #possible de rentrer les features ici
                                 ignore_index=True)



  return sub_data_collect
`````

`````python
@function_exec_time
def  extractor_nasa_data(start: int, end: int):

  """
  this function needs to install geopy module in order to get location of each
  city

  even define columns name and cities
  """

  col_name = list(city_table_response1.keys())
  col_name.insert(0, "ville")
  col_name.insert(0, "parameters")
  civ_cities = data_stock.ville
  data_collect = pd.DataFrame(columns=col_name)

  for city in civ_cities:
    lat = geolocat.geocode(city).latitude # remplacer par
    lon = geolocat.geocode(city).longitude

    new_rows = POWER_Monthly_Annual_city_data(
                                start=start,
                                end=end,
                                city=city,
                                latitude=lat,
                                longitude=lon,
                                parameters=parameters,
                                columns=col_name
                              )

    data_collect = pd.concat([data_collect, new_rows],
                            ignore_index=True)


  del lat, lon, new_rows

  return data_collect
`````
